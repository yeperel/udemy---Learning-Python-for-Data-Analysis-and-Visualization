```python
#Set array for one side of grid
points = np.arange(-5,5,0.01)
```

```python
#Create the grid
dx,dy = np.meshgrid(points,points)
```

```python
#Show what one side looks at
dx
```

```python
dy
```

```python
#Evaluating function
z = (np.sin(dx) + np.sin(dy))
```

```python
#Lets take a look at z result
z
```

```python
#Plot out the 2D array 
plt.imshow(z)
```

```python
#Plot with a colorbar
plt.imshow(z)
plt.colorbar()
```

```python
#Lets learn how to use the numpy where
#First the slow way to this
A = np.array([1,2,3,4])
B = np.array([100,200,300,400])

#Now a boolean array
condition = np.array([True,True,False,False])
#using al ist coomprehension
answer = [(A_val if cond else B_val) for A_val,B_val,cond in zip(A,B,condition)]
#Show
answer
#Problems include speed issues and multi-dimensional array issues
```

```python
#now using the numpy wehere
answer2 = np.where(condition,A,B)
#show 
answer2
```

```python
#can use np.where on 2d for manipulation
from numpy.random import randn
arr = randn(5,5)
#Show 
arr
```

```python
#Where array is less then zero, make that value zero, otherwise leave it as the array value
np.where(arr<0,0,arr)
```
